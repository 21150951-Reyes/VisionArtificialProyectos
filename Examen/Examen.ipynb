INSTRUCCIONES
Usando las herramientas vistas en clase, implemente un programa que procese imágenes de diferentes
cantidades de dados en la escena (50 %).
a.- Realice el conteo del total de dados en la imagen.
b.- Al reconocer cada dado por individual, que el programa reconozca el valor de la cara del dado
visible (número de puntos).
Realice el pre-procesamiento necesario (ajuste de brillo, contraste, morfológicas, etc.) y emplee conectividad de componentes para obtener los resultados deseados. Considere en su implementación la
escalabilidad y reusabilidad. Pruebe su código con las 15 imágenes proporcionadas. Los resultados
esperados por cada imagen se encuentran en las sub-carpeta Annotations.


```python
import cv2
import numpy as np


imagen = cv2.imread("IMGDice/Dice15.jpg")
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gris, (5,5), 0)
_, thresh = cv2.threshold(blur, 160, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Thresh", thresh)
cv2.waitKey(0)
contornos, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
dados = []
resultado = imagen.copy()
for cnt in contornos:
    area = cv2.contourArea(cnt)

    
    if 200 < area < 8000:
        x, y, w, h = cv2.boundingRect(cnt)

        
        aspecto = w / float(h)
        if 0.6 < aspecto < 1.4:
            dados.append((x, y, w, h))
            cv2.rectangle(resultado, (x,y), (x+w,y+h), (0,255,0), 2)

valores = []


for (x, y, w, h) in dados:
    roi = thresh[y:y+h, x:x+w]

    
    cont_puntos, _ = cv2.findContours(roi, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    contador = 0

    for c in cont_puntos:
        area = cv2.contourArea(c)

        
        if 30 < area < 300:
            contador += 1

    valores.append(contador)

    cv2.putText(resultado, str(contador), (x, y-10),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)


print("Cantidad de dados:", len(valores))
print("Valores:", valores)

cv2.imshow("Resultado", resultado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
